```python
import os
import shutil

dataset_path = "./yolov5/datasets/coco128"
images_path = os.path.join(dataset_path, "images/train2017")
labels_path = os.path.join(dataset_path, "labels/train2017")

filtered_images_path = os.path.join(dataset_path, "filtered/images/train2017")
filtered_labels_path = os.path.join(dataset_path, "filtered/labels/train2017")

os.makedirs(filtered_images_path, exist_ok=True)
os.makedirs(filtered_labels_path, exist_ok=True)

for label_file in os.listdir(labels_path):
    label_path = os.path.join(labels_path, label_file)

    with open(label_path, "r") as file:
        lines = file.readlines()

    person_lines = [line for line in lines if line.startswith("0 ")]

    if person_lines:
        new_label_path = os.path.join(filtered_labels_path, label_file)
        with open(new_label_path, "w") as file:
            file.writelines(person_lines)

        image_file = label_file.replace(".txt", ".jpg")
        image_path = os.path.join(images_path, image_file)
        if os.path.exists(image_path):
            shutil.copy(image_path, os.path.join(filtered_images_path, image_file))

print("Filtering complete. New dataset stored in 'filtered' directory.")
```

```python
from ultralytics import YOLO
model = YOLO("yolov5/runs/train2/weights/best.pt")
```

```python
import cv2
cap = cv2.VideoCapture("3641366019-preview.mp4")
if not cap.isOpened():
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)
    annotated_frame = results[0].plot()
    cv2.imshow("Live Camera Feed", annotated_frame)

    # press 'q' to quit the live feed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
